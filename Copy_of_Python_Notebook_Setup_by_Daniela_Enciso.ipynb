Research Question # 1: Is there an association between the amount of times an adult made you feel unwanted and avoiding your life problems? 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

```python
df['H4MA1'].value_counts()
```

```python
df['Felt_Unwanted']= df['H4MA1']
df['Felt_Unwanted'].value_counts()
```

```python
df['Felt_Unwanted'] = df['Felt_Unwanted'].loc[ df['Felt_Unwanted'] !=' ']
df['Felt_Unwanted'].value_counts()
```

```python
df['Felt_Unwanted'] = df['Felt_Unwanted'].loc[ df['Felt_Unwanted'] !='96']
df['Felt_Unwanted'] = df['Felt_Unwanted'].loc[ df['Felt_Unwanted'] !='98']
df['Felt_Unwanted'].value_counts()
```

```python
df['Felt_Unwanted'].replace(['5','3','1','2','4','6'], ['MORE THAN TEN','THREE TO FIVE','ONE','TWO','SIX TO TEN','NEVER HAPPENED'] , inplace= True)
df['Felt_Unwanted'].value_counts()
```
